### 사이킷런이란? 
   사이킷 런은 대표적인 파이썬 머신러닝 라이브러리다.
   
 <사이킷 런의 용도>
   1. Classification(분류)  
        ex)"어떤 조건의 고객에게 대출을 해줘야 하나?", "문의가 배송, 주문, 취소 중 어떤 카테고리에 해당이 되는가?"
        <p/>
   2. Refression(회귀)  
        ex) "판매량을 봤을 때 재고량을 얼마 정도 준비해 둬야하나?", "판매량을 어떻게 될까?", "매출은 얼마나 증감하게 될까?"
   <p/>
   3. Clustering (군집화) 
        ex)"고객의 군집화" 
   4. Semi-Supervised Learning
   5. Feature Selection
   6. Feature Extraction
   7. Manifold Learning
   8. Dimensionality Reduction
   9. Kernel Approximation
   10. Hyperparameter Optimization
   11. Ecaluation Metrics
   12. Out-of-core learning

#### 정답을 알고있는 문제에 대한 머신러닝: Supervised Machine Learning
     다른 말로 지도학습, 교수학습, 교사학습이라고도 불린다.
     대표적인 모델로는 RandomForestClassifier() 가 있다. 트리 계열의 의사결정방식이다.
       
      clf=RandomForestClassifier()
      clf.fit(문제 데이터, 정답 데이터)       --> 학습하는 과정!
      y_predict= clf.predict(테스트 데이터)   --> 예측하는 과정 (시험보는 과정)
      clf.score(테스트 데이터, 정답 테스트 데이터) --> 검증 과정 (정답테스트 데이터는 아무도 모른다)
     
  #### 정답을 모르는 문제에 대한 머신러닝: Unsupervised Machine Learning
      정답 데이터가 없다는 것이 지도학습과 비지도학습의 가장 큰 차이점이다.
      PCA를 통해 차원축소를 하여 만들어진 새로운 데이터를 학습시키는 것으로 진행된다.
      
      pca=PCA()
      pca.fit(x_train)
      x_new = pca.tranform(x_test)
 
  
   <estimator.fit(X)  or estimator.fit(X,Y) > ---> 비지도학습의 경우 정답데이터(Y)가 없을 것이므로 Y는 선택사항이다.<br/>
  
      cf) estimator자리에는 모델들이 들어간다 (RandomForestClassifier(),PCA()....등)
   + "estimator.pradict()" 적용 대상  
      : Classification / Regression / Clustering <p/>
      
   + "estimator.transform()" 적용 대상  
      : Preprocessing/ Dimensionality reduction/ Feature selection/ Feature extraction

```python
from sklearn import tree  #의사결정나무 라이브러리는 tree안에 있다. 
X=[[0,0],[1,1]]
Y=[0,1]  # 0과 0이 입력되면 정답은 0, 1과 1이 입력되면 정답은 1 이라는 뜻으로 학습시키고자 한다.

clf = tree.DecisionTreeClassifier()
clf
```

```python
clf=clf.fit(X,Y)  # fit= 학습한다!!  
clf.predict([[2.,2.]])  #2와 2라는 값을 넣으면 어떤 값이 나올 것인지 예측하라
```

```python
clf.predict_proba([[2.,2.]]) 
#0. , 1. 이 나왔다.... 정답이 0과 1인 것만 학습을 했으므로 1과 0으로만 값이 나온 것이다. 
#proba는 비율로 값을 출력하는 것이다.
```

## 꽃의 품종을 예측해보자

```python
from sklearn.datasets import load_iris 
   #사이키런에서 실습을 할 수 있도록 준비해둔 데이터이다.
from sklearn import tree
X,y=load_iris(return_X_y=True)

X, y
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)  #X는 정보 y는 X일때의 정의 값을 넣을 것
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
t= tree.plot_tree(clf.fit(X,y),filled=True)
```
