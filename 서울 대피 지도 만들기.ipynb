```python
import pandas as pd
```

```python
file ="./data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv"
```

```python
raw = pd.read_csv(file,encoding="cp949")
raw.head()
```

```python
raw.info()
```

```python
#0번째 인덱스에 위치한 행의 '위도' 값을 가지고 오라
i=0
lat = raw.loc[i,'위도']
long = raw.loc[i,'경도']
name = raw.loc[i,'대피소명칭']
print(lat,long,name)
```

```python
#iloc[]를 이용하면 행과 열의 인덱스 번호를 기준으로 값을 가지고 오게 된다.
raw.iloc[0,12]
```

"i" 의 값에 따라 값을 다르게 가지고 올 수 있다 i를 반복문으로 바꿔주게 되면 raw 데이터의 모든값들을 가지고 오게 할 수 있는데 
그것을 구현하기 위해서는 for range()구문을 활용할 필요가 있다

for  변수 in range() 구문은 () 안에 들어가는 숫자 갯수만큼의 인덱스를 만들어 준다.

```python
for i in range(10):
    print(i)
```

```python
# for i in range()의  () 안에 len(데이터프레임)을  넣으면 데이터프레임의 행의 갯수만큼 반복을 하게 된다.

for i in range(len(raw)):
    lat = raw.loc[i,'위도']
    long = raw.loc[i,'경도']
    name = raw.loc[i,'대피소명칭']
    print(lat,long,name)
```

```python
import folium
```

```python
folium.Map(location=[37.5536067,126.9674308],
          zoom_start=12)
```

```python
m = folium.Map(location=[37.5536067,126.9674308],zoom_start=12)

for i in range(len(raw)):
    lat = raw.loc[i,'위도']
    long = raw.loc[i,'경도']
    name = raw.loc[i,'대피소명칭']
    folium.Marker([lat,long],tooltip=name).add_to(m)  # 반복한 것을 m에 추가하라 즉 지도에 표시되게 하라
    
m
```

```python
m.save('./data/mymap.html')
```
